```python
import pandas as pd
import numpy as np
from decimal import Decimal, getcontext

# Configurar a precisão para 20 casas decimais
getcontext().prec = 50

string = "PROCESSAMENTO DE IMAGENS DIGITAIS"
#string = "BRUNO STAFUZZA MAION"
#string = "MAC ADDICT"

minimo = Decimal('0.0')
maximo = Decimal('1.0')

def reorganizarString(string):
    caracteres = sorted(set(string))
    if ' ' in caracteres:
        caracteres.remove(' ')
        caracteres.insert(0, ' ')
    return ''.join(caracteres)

stringOrdem = reorganizarString(string)


def contarCharString(string, c):
    count=0
    for i in string:
        if i==c:
            count=count+1
    return count

char = " "

#print(contarCharString(string, char))

lista_elementos = []
stringLen = len(string)
valorCadaElemento = maximo/stringLen


for i in stringOrdem:
    elementoReptido = contarCharString(string,i)
    lista_elementos.append(elementoReptido)

lista_min = []
lista_max = []

novo_min = minimo
for i, e in enumerate(lista_elementos):
    #lista_min.append(int(minimo2))
    lista_min.append((novo_min))
    novo_min = novo_min + (lista_elementos[i] * valorCadaElemento)
    #lista_max.append(int(minimo2))
    lista_max.append((novo_min))

string_lista = list(stringOrdem)
dfProbabilidade = pd.DataFrame({
    'CARACTERES': string_lista,
    'OCORRENCIAS': lista_elementos,
    'MIN': lista_min,
    'MAX': lista_max,
})

display(dfProbabilidade)
```

# CODIFICAÇÃO

```python


novo_min = minimo
novo_max = maximo

lista_intervalo = []
lista_min = []
lista_max = []

for i, e in enumerate(string):
  lista_intervalo.append(e)
  linha = dfProbabilidade[dfProbabilidade['CARACTERES'] == e].iloc[0]
  minLinha = linha['MIN']
  maxLinha = linha['MAX']

  auxNovoMin = novo_min#pra nao mudar o calculo
  novo_min = novo_min + (novo_max-novo_min)*minLinha
  novo_max = auxNovoMin + (novo_max-auxNovoMin)*maxLinha


  lista_min.append(novo_min)
  lista_max.append(novo_max)

print(lista_min)

dfCodificacao = pd.DataFrame({
    'CARACTER': lista_intervalo,
    'MIN': lista_min,
    'MAX': lista_max,
})

display(dfCodificacao)
numeroCodifica = lista_min[-1]
print("NÚMERO CODIFICADO",numeroCodifica)
```

# DESCODIFAÇÃO


```python
lista_numero = []
lista_simbolo = []
lista_min = []
lista_max = []
lista_delta = []

for i, e in enumerate(string):
  lista_numero.append(numeroCodifica)
  #ENCONTRA O INTERVALO
  for idx, row in dfProbabilidade.iterrows():
      if row['MIN'] <= numeroCodifica < row['MAX']:
          linha = row
          break

  minimo = linha['MIN']
  maximo = linha['MAX']
  lista_min.append(minimo)
  lista_max.append(maximo)

  delta = maximo-minimo
  lista_delta.append(delta)

  simbolo = linha['CARACTERES']
  lista_simbolo.append(simbolo)

  numeroCodifica = (numeroCodifica-minimo)/delta


  

dfDescodificacao = pd.DataFrame({
    'NUMERO': lista_numero,
    'SIMBOLO': lista_simbolo,
    'MIN': lista_min,
    'MAX': lista_max,
    'DELTA': lista_delta,
})



nova_string = ''
for i in lista_simbolo:
   nova_string = nova_string + i

print("STRING DESCODIFICADA:", nova_string)

display(dfDescodificacao)

```
